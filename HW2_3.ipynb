##2.3.1 Joint Probability Distributions

key terms:

Joint Probabilty: The probability that two or more events occur at the same time.

Joint Probability Distribution: Probability distribution of multiple variables.

For discrete variables X and Y, the JPD is represented as:

p(x, y) = P(X = x and Y = y)


For continuous variables X and Y, the JPD is represented as:

P[(X,Y) ∈ A] = ∫∫f(x,y)dxdy

```python
#Example of Joint Probability DIstribution
import numpy as np
import matplotlib.pyplot as plt
import math

print("Suppose X and Y are continuous variables which represent the height and weight of an individual\n Let the mean of X = 1.75m and standard deviation of X = 0.1m\n Let the mean of Y = 70 kg and standard deviation of Y = 10kg\n Let the coorelation coefficient equal 0.7")


mean = [1.75, 70]  #array of means of x,y (needed for numpy calculations)
stdX = 0.1
stdY = 10
p = 0.7

cov = np.array([[math.pow(stdX,2), p * stdX * stdY],[p * stdX * stdY, math.pow(stdY,2)]]) #Covarience matrix
print("Covarience Matrix (represents variences in X and Y):")
print(cov)

data = np.random.multivariate_normal(mean, cov, 1000) #method to generate samples for a nomral distribution
height = data[:,0] #gathering height and weight data from data matrix
weight = data[:,1]

plt.figure(figsize=(5, 5)) #plotting height v weight data
plt.scatter(height, weight, 0.5)
plt.xlim(0, 3)
plt.title('Joint Probabiltiy Distribution Height v Weight')
plt.xlabel('Height (m)')
plt.ylabel('Weight (kg)')
plt.grid()
plt.show()

print("The areas with larger cluster of datapoints represent a higher probability that an indiivdual is within that range of height and weights.\n Similarly, areas with fewer points represent a lower probability that an individual is within that range of heights and weights.")
```

##2.3.3 Random Samples

A completely random subset of people from a larger set of people to observe trends in the set.

A sample is considered random if it satisfies the two properties:

1. The Xn ’s are independent random variable’s .
2. Every Xn has the same probability distribution.

Central Limit Theorem:

The distribution of a sample approaches a normal distribution as the sample size grows
